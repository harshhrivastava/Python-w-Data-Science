```python
from dputils.scrape import Scraper, Tag
import pandas as pd

# The Scraper class is a wrapper around BeautifulSoup that makes it easier to scrape web pages.
# It provides a few convenience methods and a few convenience attributes.
# Tag is a subclass of BeautifulSoup's Tag class that provides a few convenience methods.

# We need to reverse engineer in order to mine the data we want.
# The target area is the area that contains the data we want.

url = 'https://www.zyte.com/blog/'
target = Tag(cls='ResourcePage_grid__avGjs')
items = Tag(cls='CardResource_card__BhCok')
article = Tag(cls='z-text-blog-title')
author = Tag('div', cls='z-text-small')
link = Tag('a', cls='gtm-linkclick', output='href')

scr = Scraper(url)
data = scr.get_repeating_page_data(target, items, article=article, link=link, author=author)

dataframe = pd.DataFrame(data)

# From here on we can continue with data analysis.
# We will be doing the same for Amazon.

data = []
page = 1
k = 'smartphone'
target = Tag(cls='s-search-results')
items = Tag(cls='s-result-item')
name = Tag('h2')
price = Tag('span', cls='a-offscreen')
original_price = Tag('span', cls='a-price a-text-price')
reviews = Tag('span', cls='a-icon-alt')
for i in range(1, 50):
    url = f'https://www.amazon.in/s?k={k}&page={page}'
    scr = Scraper(url)
    results = scr.get_repeating_page_data(
        target,
        items,
        name = name,
        price = price,
        original_price = original_price,
        reviews = reviews
    )
    if len(results) == 0:
        print(f'No more results found on page {page}.')
        break
    data.extend(results)
    page += 1

dataframe = pd.DataFrame(data)
dataframe['original_price'] = dataframe['original_price'].str.split('₹').str[1]
dataframe['price'] = dataframe['price'].str.split('₹').str[1]
dataframe.dropna(inplace=True)
dataframe['reviews'] = dataframe['reviews'].str.split(' ').str[0]
dataframe['original_price'] = dataframe['original_price'].str.replace(',', '')
dataframe['price'] = dataframe['price'].str.replace(',', '')
dataframe['name'] = dataframe['name'].str.strip()
print(dataframe)
dataframe.to_csv('amazon.csv')
```
