### Built-in methods in python

- print
- input
- eval
- exec
- round
- type
- minmap
- abs
- isinstance
- all
- len
- max
- min
- sum
- any
- open

```python
# The print function is used to print the output to the console.
# It can take any number of arguments.
# The optional arguments end, sep and file can be used to change the default behaviour of the print function.
# The end argument is used to change the default newline character.
# The sep argument is used to change the default separator between the arguments.
# The file argument is used to change the default output stream.
print('Hello World') # Hello World
print('Hello', 'World', sep=' ') # Hello World
print('Hello', 'World', sep=' ', end='\n') # Hello World
print('This will be added', 'to the dummy text file', sep=' ', end='', file=open('./exercises/dummy.txt', 'w')) # Hello World

# The input function is used to take input from the user.
# It takes one optional argument which is the prompt message.
# The input function returns the input as a string.
name = input('Enter your name: ')
print('Hello', name, sep=' ')

# The exec function is used to execute a string as python code.
# It takes one argument which is the string to be executed.
# The exec function returns None.
# Difference between exec function and eval function is that exec function does not return anything.
exec('print("Hello World")')

# The type function is used to find the type of an object.
# It takes one argument which is the object whose type is to be found.
# The type function returns the type of the object.
print(type(1)) # <class 'int'>

# The map function is used to apply a function to each element of an iterable.
# It takes two arguments, the first argument is the function to be applied
# and the second argument is the iterable.
# The map function returns a map object which can be converted to a list.
def square(x):
    return x*x
l = [1, 2, 3, 4, 5]
print(list(map(square, l))) # [1, 4, 9, 16, 25]

# The len function is used to find the length of an iterable.
# It takes one argument which is the iterable whose length is to be found.
# The len function returns the length of the iterable.
print(len(l)) # 5

# The sum function is used to find the sum of all the elements of an iterable.
# It takes one argument which is the iterable whose sum is to be found.
# The sum function returns the sum of all the elements of the iterable.
print(sum(l)) # 15

# The isinstance function is used to check whether or not an object is an instance of a particular class.
# It takes two arguments, the first argument is the object and the second argument is the class.
# It returns true if the object is an instance of the class else it returns false.
print(isinstance(1, int)) # True
print(isinstance(1.0, int)) # False

# The abs function is used to find the absolute value of a number.
# It takes one argument which is the number whose absolute value is to be found.
print(abs(-1)) # 1

# The round function is used to round off a number to a particular number of digits.
# It takes two arguments, the first argument is the number to be rounded off
# and the second argument is the number of digits.
# If no second argument is provided then the number is rounded off to the nearest integer.
print(round(1.23456789, 2)) # 1.23

# Eval function is used to evaluate an expression which is in the form of a string.
a = 10
b = 20
print(eval('a+b')) # This will evaluate to the sum of variables a and b.

# All function is a validation function which checks whether or not all the elements
# of an iterable complies to a particular condition or not.
# If all the elements of an iterable are true then all function returns true.
# If any of the element of an iterable is false then all function returns false.
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 0]
print(all(x)) # True
print(all(y)) # False

# Any function is a validation function which checks whether or not any of the elements
# of an iterable complies to a particular condition or not.
# If any of the element of an iterable is true then any function returns true.
# If all the elements of an iterable are false then any function returns false.
z = [0, 0, 0, 0, 0]
print(any(x)) # True
print(any(y)) # True
print(any(z)) # False

# The max function is used to find the maximum value from an iterable.
l1 = [1, 2, 3, 4, 5]
l2 = [1, 2, 3, 4, 0]
print(max(l1)) # This will print 5.

# The min function is used to find the minimum value from an iterable.
print(min(l1)) # This will print 1.
```
